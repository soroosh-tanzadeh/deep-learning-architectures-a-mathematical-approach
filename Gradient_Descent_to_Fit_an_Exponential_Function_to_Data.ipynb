# Gradient Descent to Fit an Exponential Function to Data

## Problem Statement

Given a set of data points $(x_1, y_1), (x_2, y_2), \dots, (x_n, y_n)$, we want to fit an exponential function of the form:

$$
f(x) = ae^{bx}
$$

where \( a \) and \( b \) are the parameters we need to find. The goal is to use **Gradient Descent** to minimize the **Mean Squared Error (MSE)** between the actual data points and the predicted values from the exponential function.

## Steps

### 1. Define the Model

The exponential function we want to fit is:

$$
f(x) = ae^{bx}
$$

where:
-  $a$ and $b$ are the parameters that we need to determine.
- $x$ is the input variable.
  
### 2. Define the Cost Function

The goal is to minimize the **Mean Squared Error (MSE)** between the actual data points $(x_i, y_i)$ and the predicted values from the exponential function $f(x_i)$:

$$
C(a, b) = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - ae^{b x_i} \right)^2
$$

Where:
- $y_i$ is the actual value at point $x_i$,
- $ae^{b x_i}$ is the predicted value from the exponential function.

### 3. Compute the Gradients

To use **Gradient Descent**, we need to compute the partial derivatives of the cost function with respect to the parameters $a$ and $b$.

### 4. Gradient Descent Update Rule

We use the **Gradient Descent** update rule to iteratively update $a$ and $b$ to minimize the cost function:

$$
a \leftarrow a - \eta \frac{\partial C}{\partial a}
$$

$$
b \leftarrow b - \eta \frac{\partial C}{\partial b}
$$

Where $\eta$ is the learning rate, a small positive number that determines how large each update step should be.

### 5. Set Initial Values

- Start with initial guesses for $a$ and $b$, e.g., $a = 1$ and $b = 0$.
- Choose a learning rate $\eta$, e.g., $\eta = 0.01$.
- Set a number of iterations or a threshold for stopping.

### 6. Iterate

Use the gradient descent update rules to update $a$ and $b$ for a set number of iterations or until the changes in $a$ and $b$ become sufficiently small.

# Data Preparation

```python
import numpy as np
import matplotlib.pyplot as plt

true_a = np.random.uniform(0.5, 2.0)
true_b = np.random.uniform(0.1, 0.5)

print(f"True parameters: a = {true_a:.2f}, b = {true_b:.2f}")

x_data = np.linspace(0, 5, 50)

y_true = true_a * np.exp(true_b * x_data)

noise_std = 0.5
noise = np.random.normal(0, noise_std, len(x_data))
y_data = y_true + noise

plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, label='Noisy Data', color='blue', s=20)
plt.plot(x_data, y_true, label=f'True Function: {true_a:.2f}e^({true_b:.2f}x)', color='red', linestyle='--')
plt.title('Generated Noisy Exponential Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

data = np.column_stack((x_data, y_data))
```
